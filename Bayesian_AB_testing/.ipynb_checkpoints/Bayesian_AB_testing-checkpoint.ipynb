# Bayesian A/B Testing with PyMC

This notebook is a demonstration of how you can use the Python PyMC library to conduct a *Bayesian* A/B test on website content.  The Bayesian approach is ideal because it provides insight on our data *given the data we have* (in other words, we can't know the full population).  

Using some mock web traffic data, I've built this A/B test around conversion rate for users who saw different images on the website:
- A grapefruit
- A durian fruit

The population is split into two equal groups (50/50 split).  At random, 50% of site visitors saw the grapefruit, while the other 50% saw the durian fruit.  

## Testing and Analysis


```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import pylab as pl
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')

%matplotlib inline
```

Loading the data from our test.

```python
data = pd.read_csv('web_data.csv')
```

```python
data.head()
```

```python
# Group 1 are people who saw the grapefruit image; group 2 are people who saw the durian fruit image.
group1 = data[data.Group == 0]
group2 = data[data.Group == 1]
```

```python
g1samp = group1['Clicks Conv'].tolist()
g2samp = group2['Clicks Conv'].tolist()
```

```python
sns.distplot(g1samp, axlabel='Conversions')
plt.title('Users Who Saw Grapefruit Image')
```

```python
sns.distplot(g2samp, axlabel='Conversions')
plt.title('Users Who Saw Durian Fruit Image')
```

These groups appear quite different.  Group 2's mean (~22) is much higher than group 1 (~12).  The next step is to calculate the credible intervals for the mean for both both groups.  

```python
group1_bayes = stats.bayes_mvs(g1samp, alpha=0.95)
```

```python
group1_bayes
```

```python
group2_bayes = stats.bayes_mvs(g2samp, alpha=0.95)
```

```python
group2_bayes
```

This is informative and suggests that Group 2 (durian fruit) performed better, since its mean conversions was higher.  However, this output doesn't make it entirely clear whether the difference in means is meaningful.

In order to come to a more informed decision, we will conduct a Bayesian estimation to determine *how* different the groups are, using the Markov Chain Monte Carlo method to estimate priors.

```python
import pymc3 as pm
```

```python
mean_prior_mean = data['Clicks Conv'].mean()
mean_prior_std = data['Clicks Conv'].std()
```

```python
# Setting prior as Normal distribution, since conversion rates are continous, and the histograms plotted 
# above look quite close to normal
with pm.Model() as model:

    group1_mean = pm.Normal('Group1_mean', mean_prior_mean, sd=mean_prior_std)
    group2_mean = pm.Normal('Group2_mean', mean_prior_mean, sd=mean_prior_std)
```

```python
#Determining the lowest min value and highest max value across the two sample groups in order to define the range
#for Standard Deviation
print "G1 min: " + str(min(g1samp))
print "G2 min: " + str(min(g2samp))
print "G1 max: " + str(max(g1samp))
print "G2 max: " + str(max(g2samp))
```

```python
std_prior_lower = 3.0
std_prior_upper = 39.0

with model:
    
    group1_std = pm.Uniform('Group1_std', lower=std_prior_lower, upper=std_prior_upper)
    group2_std = pm.Uniform('Group2_std', lower=std_prior_lower, upper=std_prior_upper)
```

```python
# Setting observed values

with model:

    group1 = pm.Normal('Group1', mu=group1_mean, sd=group1_std, observed=g1samp)
    group2 = pm.Normal('Group2', mu=group2_mean, sd=group2_std, observed=g2samp)
```

```python
# Setting up key comparitive metrics

with model:

    diff_of_means = pm.Deterministic('difference of means', group1_mean - group2_mean)
    diff_of_stds = pm.Deterministic('difference of stds', group1_std - group2_std)
    effect_size = pm.Deterministic('effect size',
                                   diff_of_means / np.sqrt((group1_std**2 + group2_std**2) / 2))
```

```python
# Using MCMC to fit the model
with model:
    trace = pm.sample(25000, njobs=4)
```

```python
# Plotting the posteriors
pm.plot_posterior(trace[3000:],
                  varnames=['Group1_mean', 'Group2_mean', 'Group1_std', 'Group2_std'],
                  color='#87ceeb')
```

```python
# Plotting diefferences of means and stds
pm.plot_posterior(trace[3000:],
                  varnames=['difference of means', 'difference of stds', 'effect size'],
                  ref_val=0,
                  color='#87ceeb')
```

## Interpretation
The charts above make abundtantly clear that there is a very significant difference in average conversions between the group that saw the grapefruit and the group that saw the durian fruit.  Not only is Group 2's (durian fruit) mean much higher than Group 1, the credible intervals for both group's means do not overlap.  (If they did, then the green line in the "difference of means" graph above would overlap with some of the blue lines.)

Given this dramatic difference, the recommendation would clearly be to implement the durian fruit image on the site.
